```python
def dfs(cap1, cap2, target):
    visited = set()
    stack = [((0, 0), [])]

    while stack:
        (x, y), path = stack.pop()

        if (x, y) in visited:
            continue

        visited.add((x, y))
        path = path + [((x, y), "Start" if not path else path[-1][1])]

        if x == target or y == target:
            print("Goal Reached:\n")
            for state, rule in path:
                print(f"State: {state}  | Rule: {rule}")
            return

        next_states = []

        next_states.append(((cap1, y), "Fill Jug1"))
        next_states.append(((x, cap2), "Fill Jug2"))
        next_states.append(((0, y), "Empty Jug1"))
        next_states.append(((x, 0), "Empty Jug2"))

        transfer = min(x, cap2 - y)
        next_states.append(((x - transfer, y + transfer), "Pour Jug1 -> Jug2"))

        transfer = min(y, cap1 - x)
        next_states.append(((x + transfer, y - transfer), "Pour Jug2 -> Jug1"))

        for state, rule in next_states:
            if state not in visited:
                stack.append((state, path + [(state, rule)]))

    print("No Solution Found")

dfs(4, 3, 2)
```
