# OpenCV Main Functions Notebook
This notebook demonstrates core OpenCV operations for image processing.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
print('OpenCV version:', cv2.__version__)
```

## Load Image

```python
img = cv2.imread('sample.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.title('Original Image')
plt.axis('off')
```

## Convert to Grayscale

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.title('Grayscale')
plt.axis('off')
```

## Resize Image

```python
resized = cv2.resize(img_rgb, (300,300))
plt.imshow(resized)
plt.title('Resized')
plt.axis('off')
```

## Edge Detection

```python
edges = cv2.Canny(gray,100,200)
plt.imshow(edges, cmap='gray')
plt.title('Edges')
plt.axis('off')
```

## Draw Shapes

```python
shape_img = img_rgb.copy()
cv2.rectangle(shape_img,(50,50),(200,200),(255,0,0),3)
cv2.circle(shape_img,(300,150),50,(0,255,0),3)
plt.imshow(shape_img)
plt.title('Shapes')
plt.axis('off')
```

## Text on Image

```python
text_img = img_rgb.copy()
cv2.putText(text_img,'OpenCV',(50,100),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,0),2)
plt.imshow(text_img)
plt.title('Text')
plt.axis('off')
```

## Blur Image

```python
blur = cv2.GaussianBlur(img_rgb,(15,15),0)
plt.imshow(blur)
plt.title('Blur')
plt.axis('off')
```

## Save Image

```python
cv2.imwrite('output.jpg', cv2.cvtColor(blur, cv2.COLOR_RGB2BGR))
print('Saved as output.jpg')
```
