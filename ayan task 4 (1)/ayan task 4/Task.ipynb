# Task 1; Implement the N-Queens Problem (Dynamic)

```python
def solve_n_queens(n):
    board = [-1] * n
    solutions = []

    def is_safe(row, col):
        for i in range(row):
            if board[i] == col or abs(board[i] - col) == abs(i - row):
                return False
        return True

    def dfs(row):
        if row == n:
            solutions.append(board[:])
            return
        for col in range(n):
            if is_safe(row, col):
                board[row] = col
                dfs(row + 1)
                board[row] = -1

    dfs(0)
    return solutions


def print_solutions(solutions, n):
    for idx, sol in enumerate(solutions, 1):
        print(f"\nSolution {idx}:")
        for i in range(n):
            row = ["Q" if sol[i] == j else "." for j in range(n)]
            print(" ".join(row))
    print(f"\nTotal Solutions: {len(solutions)}")


n = int(input("Enter value of N: "))
solutions = solve_n_queens(n)
print_solutions(solutions, n)
```
